```python
import networkx as nx

G = nx.Graph()
```

```python
kite_network = [
    ("Carol", "Andre"), ("Carol", "Diane"), ("Carol", "Fernando"),
    ("Andre", "Beverly"), ("Andre", "Diane"), ("Andre", "Fernando"),
    ("Fernando", "Diane"), ("Fernando", "Garth"), ("Fernando", "Heather"),
    ("Diane", "Beverly"), ("Diane", "Ed"), ("Diane", "Garth"),
    ("Heather", "Garth"), ("Heather", "Ike"),
    ("Ike", "Jane"),
    ("Beverly", "Ed"), ("Beverly", "Garth"),
    ("Garth", "Ed")
    ]

pos = {
    "Carol": [2, 5],
    "Andre": [1, 4],
    "Fernando": [3, 4],
    "Diane": [2, 3],
    "Heather": [4, 3],
    "Ike": [5, 3],
    "Jane": [6, 3],
    "Beverly": [1, 2],
    "Garth": [3, 2],
    "Ed": [2, 1]
}

G.add_edges_from(kite_network)
```

```python
nx.draw(G, pos=pos, with_labels=True)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```
